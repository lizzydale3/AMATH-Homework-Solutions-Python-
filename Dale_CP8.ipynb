```python
import numpy as np
import scipy.integrate
######################

### Problem 1
### Initialize t, and x_true
x_true = lambda t: 0.5*(np.cos(t) + np.sin(t) + np.e**(-t))
dt = 0.1
T = 10
t = np.arange(0,T+dt, dt)
n = t.size
x = np.zeros(n)
x[0] = 1

### Forward Euler
### Write a forward Euler scheme
for i in range (n-1):
  x[i+1] = x[i] + dt*(np.cos(t[i])-x[i])

A1 = np.reshape(x.copy(), (1,101))
err = np.zeros(n)

for j in range(n):
  err[j] = np.abs(x[j]-x_true(t[j]))

A2 = err.copy().reshape(1,101)


### Backward Euler
### Write a backward Euler scheme
dt = 0.1
T = 10
t = np.arange(0,T+dt, dt)
n = t.size
x = np.zeros(n)
x[0] = 1

for k in range(n - 1):
    x[k + 1] = ((x[k] + dt * np.cos(t[k+1])) / (1+dt))

A3 = x.copy().reshape(1, 101)
err2 = np.zeros(n)

for m in range (n):
  err2[m] = np.abs(x[m] - x_true(t[m]))

A4 = err2.copy().reshape(1,101)

### Built-in Solver
### Use scipy.integrate.solve_ivp
### Don't forget to reshape the solution you get from scipy.integrate.solve_ivp

f = lambda t, x: np.cos(t) - x
T = 10
t = np.arange(0, T + dt, dt)
tspan = (0, T)
t_eval = np.arange(0, 10 + 0.1, 0.1)
x_0 = np.array([1])
n = t.size

sol = scipy.integrate.solve_ivp(f, tspan, x_0, t_eval=t)
x = sol.y[0, :]
A5 = np.reshape(x.copy(), (1, 101))
err3 = np.zeros(n)

for p in range(n):
    err3[p] = np.abs(x[p] - x_true(t[p]))

A6 = err3.copy().reshape(1, 101)
```
